```python
import pandas as pd

# تحميل الملف
data=pd.read_csv(r'C:\Users\mohda\OneDrive\Desktop\car price prediction\data\car_price_prediction.csv')
# تحديد الأعمدة الرقمية فقط
data_numeric = data.select_dtypes(exclude='object')

# إنشاء mask يحتفظ بالصفوف غير الشاذة
mask = pd.Series(True, index=data.index)

for col in data_numeric.columns:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    low = q1 - 1.5 * iqr
    high = q3 + 1.5 * iqr

    outliers = ((data[col] < low) | (data[col] > high)).sum()
    total = data[col].shape[0]
    print(f"Total Outliers in '{col}': {outliers} ({round(100 * outliers / total, 2)}%)")

    # تحديث القناع ليحتفظ بالصفوف المقبولة فقط
    mask &= (data[col] >= low) & (data[col] <= high)

# حذف جميع الصفوف الشاذة دفعة واحدة
data_cleaned = data[mask].reset_index(drop=True)
```

## 4-Model

```python
x=data.drop('Price',axis=1)
y=data['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
Algorithm=['LinearRegression','DecisionTreeRegressor','RandomForestRegressor',
'GradientBoostingRegressor','XGBRegressor','SVR']
R2=[]
RMSE=[]
```

```python
def models(model):
    model.fit(x_train, y_train)
    pre=model.predict(x_test)
    r2=r2_score(y_test,pre)
    rmse=np.sqrt(mean_squared_error(y_test,pre))
    R2.append(r2)
    RMSE.append(rmse)
    score=model.score(x_test,y_test)
    print(f" The Score of Model is : {score}")
```

```python
model1=LinearRegression()
model2=DecisionTreeRegressor()
model3=RandomForestRegressor()
model4=GradientBoostingRegressor()
model5=XGBRegressor()
model6=SVR()
```

```python
models(model1)
models(model2)
models(model3)
models(model4)
models(model5)
models(model6)
```

```python
df=pd.DataFrame({'Algorithm':Algorithm,"R2_Score":R2,'RMSE':RMSE})
```

```python
import matplotlib.pyplot as plt

# تأكد أن df يحتوي على الأعمدة ['Algorithm', 'R2_Score']
fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(df['Algorithm'], df['R2_Score'], label='R² Score', marker='v', color='blue')
ax.set_title('R² Scores by Algorithm')
ax.set_xlabel('Algorithm')
ax.set_ylabel('R² Score')
ax.grid(True)
ax.legend()
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
